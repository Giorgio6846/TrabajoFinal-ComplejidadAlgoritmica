```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx


df = pd.read_csv('../Generador/Dataframe.csv')
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df1 = df.loc[:,["direccion","departamento","provincia","distrito"]]
df1
```

```python

#G = networkx.Graph()

G = nx.from_pandas_edgelist(df1, "direccion", "distrito")
G

```

```python
Distritos = list(df1["distrito"].unique())
Distritos   
```

```python
Departamentos = list(df1["departamento"].unique())
Departamentos
```

```python
dict(zip(Distritos, Distritos))
```

```python

GFinal = nx.from_pandas_edgelist(
    df1, source="departamento", target="distrito")
pos = nx.draw_circular(GFinal)
#GFinal = nx.from_pandas_adjacency(df1)

plt.figure(3, figsize=(12, 12))
nx.draw(GFinal, pos, with_labels=True)
plt.show()
```

```python
Departamentos = list(df1["departamento"].unique())
Departamentos

for index in range(len(Departamentos)):
    departamento = df1.loc[df1["departamento"] == Departamentos[index]]
    GTemp = nx.from_pandas_edgelist(
        departamento, source="direccion", target="distrito")
    GFinal = nx.compose(GFinal, GTemp)
```

```python
plt.figure(3, figsize=(12, 12))
nx.draw(GFinal, pos, with_labels=False)
plt.show()
```
