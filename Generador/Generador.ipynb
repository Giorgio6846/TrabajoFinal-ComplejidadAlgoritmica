```python
import pandas as pd
from faker import Faker
import random
from datetime import date, timedelta

df = pd.read_excel("UBIGEODISTRITOS.XLSX")
df
```

```python
ListaDepartamentos = df['NOMBDEP'].unique().tolist()
ListaDepartamentos
```

```python
CodigosUbigeo = df['IDDIST'].unique().tolist()
CodigosUbigeo
```

```python
fake = Faker('es_ES')

size = 50

def row(num):
        vivienda = num
        direccion = fake.street_name() + str(" ") + str(random.randint(101, 1000))
        tipoVivienda = random.choice(['Casa', 'Apartamento', 'Departamento'])
        material_paredes = random.choice(['Ladrillo', 'Concreto', 'Adobe'])
        material_techo = random.choice(['Teja', 'Techo de Paja', 'Eternit'])
        material_piso = random.choice(['Ceramica', 'Parquet', 'Tierra'])
        cuartos = random.randint(1, 6)
        banos = random.randint(1, 4)
        cocinas = random.randint(1, 2)
        comedores = random.randint(0, 2)
        dormitorios = random.randint(1, cuartos)
        personas_vivienda = random.randint(1, 10)
        jefe_hogar = fake.name()
        ubigeo = random.choice(CodigosUbigeo)
        departamento = df.loc[df['IDDIST'] == ubigeo, 'NOMBDEP'].item()
        provincia = df.loc[df['IDDIST']==ubigeo, 'NOMBPROV'].item()
        distrito = df.loc[df['IDDIST'] == ubigeo, 'NOMBDIST'].item()
        date_time = fake.date_time()
        fecha_censo = (date(2017, 7, 1) + timedelta(days=random.randint(0, 30))).strftime("%d/%m/%Y")
        quien_respondio = jefe_hogar 
    
        data = {
            "vivienda" : vivienda,
            "direccion" : direccion,
            "tipoVivienda" : tipoVivienda,
            "material_paredes" : material_paredes,
            "material_techo" : material_techo,
            "material_piso" : material_piso,
            "cuartos" : cuartos,
            "banos" : banos,
            "cocinas" : cocinas,
            "comedores" : comedores,
            "dormitorios" : dormitorios,
            "personas_vivienda" : personas_vivienda,
            "jefe_hogar" : jefe_hogar,
            "departamento" : departamento,
            "provincia" : provincia,
            "distrito": distrito,
            "date_time" : date_time,
            "ubigeo" : ubigeo,
            "fecha_censo" : fecha_censo,
            "quien_respondio": quien_respondio
        }
    
        return data

def create_rows_faker(num=1):
    output = [row(x) for x in range(num)]

    return output

df_faker = pd.DataFrame(create_rows_faker(50000))
df_faker.to_csv('Dataframe.csv')
df_faker = df_faker.drop(['vivienda'], axis = 1)
df_faker
```
